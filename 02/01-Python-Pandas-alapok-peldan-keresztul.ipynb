## Hiányzó értékek kezelése

Ahogy a fenti táblában is láthatjuk, a Pandas NaN-al jelöli meg azokat a mezőket, ahova nem tartozik érték. Ezt még szoktuk null-nak, na-nak, vagy None-nak is hívni. Ezek helyettesítésével nagyon óvatosan kell bánnunk, hiszen ezek a későbbi gépi tanuló modellünket "torzítani" tudják. Például nem mindegy, hogy a hiányzó évszámokat 0-val, az évszámok átlagával, vagy pl. idei év - 5-el helyettesítjük. 
Ha sok adatunk van, és az adatstruktúrából adódóan megengedhető, akkor a legbiztosabb, ha minden olyan sort eldobunk, amiben NaN érték van. Ezt a gyakorlatban azonban szinte sosem tudjuk megtenni. Most mégis, menjünk így tovább.
Előbb nézzük meg, hogy hány ilyen sor van:

```python
print ("%d (sorok, amiben van NaN) / %d (összes)" % (len(df[df.isnull().any(axis=1)]), len(df)))
# a df.isnull() az eredetivel azonos méretű táblát ad vissza, mely celláiban booleanok mutatják, hogy NaN-e az érték, vagy nem.
# az .any(axis=1) ebből soronként (az axis=1 miatt) vizsgálja meg, hogy bármelyik érték NaN-e.
```

Majd tartsuk csak meg azokat, melyekben egyáltalán nincs NaN:

```python
df = df[~df.isnull().any(axis=1)] # a ~ jel a negálás operátor, ami működik DataFrame szinten is.
print("Új hossz: ", len(df))
```

*Megjegyzés: több lehetőség is van a változók megjelenítésére a print fügvénnyel, ezekről bővebben itt lehet olvasni: https://pyformat.info/*

## Kategorikus változók

A gépi tanuló eljárások számára jó, ha a kategórikus változókat külön kezeljük a numerikus változóktól. Ilyen például az, hogy a film színes vagy feketefehér. Ahhoz, hogy melyek a kategorikus változók, nézzünk bele ismét a DataFrame-be:

```python
df.head(5)
```

"Első ránézésre" az alábbi tulajdonságok tűnnek kategórikus változóknak: 

    color
    director_name
    actor_1_name
    actor_2_name
    actor_3_name
    language
    country
    content_rating

további feldolgozást kíván, de kategorikus változót lehet belőle csinálni: 

    genres
    plot_keywords
    
Vizsgáljuk meg ezeket részletesebben:

```python
categorical_features = ["color","director_name","actor_1_name","actor_2_name","actor_3_name","language","country","content_rating"]
n_items = len(df)
for feat in categorical_features:
    print("A %s feldolgozása. Egyedi mezők száma: %d" % (feat, df[feat].nunique()))
    if df[feat].nunique()<50: # ha kezelhető mennyiségű, kiírjuk ezeket
        print(df[feat].unique())
        print("Az egyedi értékek darabszáma:")
        print(df.groupby(feat).count().iloc[:,0]) # a feat-ra csoportosítunk, majd megszámoljuk, hogy hány elem van a csoportban, és kiválasztjuk az iloc-al az első oszlopot (hiszen minden oszlopba beleírta ezt az értéket)
        # összevonjuk azokat a kategóriákat, ahol az össz darabszám kevesebb, mint 2% található
        for feat_value in df[feat].unique():
            if (len(df[df[feat]==feat_value]) / n_items <= 0.02):
                print("A %s kategória hozzáadása az OTHER kategóriához." % feat_value)
                df[feat] = df[feat].apply(lambda x: "OTHER" if x==feat_value else x, 1)
                
        print("Az egyedi értékek végső darabszáma:")
        print(df.groupby(feat).count().iloc[:,0]) # a feat-ra csoportosítunk, majd megszámoljuk, hogy hány elem van a csoportban, és kiválasztjuk az iloc-al az első oszlopot (hiszen minden oszlopba beleírta ezt az értéket)
        print("\n")
        
            
```

Látjuk, hogy a color, language és content_rating esetén könnyű a dolgunk, ezeket un. one-hot vektorokká alakítjuk a Pandas következő függvényével:

```python
df = pd.get_dummies(df, columns=["color","language","country","content_rating"])
```

```python
df.head()
```

### Genres és plot_keywords tulajdonságok

Ezekben az esetekben előbb szét kell bontanunk a |-al elválasztott stringeket szavakra, és ezek számosságát kell megvizsgálnunk. 

```python
genres = pd.Series()

for value in df['genres']:
    genres = genres.append(pd.Series(value.split("|")), ignore_index=True)
    
genres = genres.drop_duplicates()
print(genres)
```

Szerencsére ez kezelhető mennyiségű kategória. Ennyi új tulajdonságot fogunk létrehozni a DateFrame-ünkben, és soronként 1-re állítjuk, amelyik szerepel a Genres-ben, és 0-ra, amelyik nem.

```python
for feat in genres:
    df[feat] = 0 # minden sornál nullára állítjuk az értékeket
    df[feat] = df['genres'].apply(lambda x: 1 if feat in str(x) else 0, 1)
    print("%s típusú filmek száma:\t %d / %d" % (feat,len(df[df[feat]==1]), len(df)))
```

## Outlierek keresése

A gépi tanuló algoritmusok számára gyakran problémát jelentenek az ún. outlierek az adatban. Ezek lehetnek a folyamatból adódó outlierek (anomáliák), de lehetnek adatbázis hibák is. Most a következőképp keressük ezeket: ha a várható érték +/- szórás háromszorosánál kisebb vagy nagyobb érték van az adatbázisban, akkor azt a szélsőértékkel helyettesítjük:

```python
import numpy as np

numeric_features = ['num_critic_for_reviews', 
                    'duration',
                    'director_facebook_likes', 
                    'actor_3_facebook_likes', 
                    'actor_1_facebook_likes', 
                    'gross', 
                    'num_voted_users', 
                    'cast_total_facebook_likes',
                    'facenumber_in_poster', 
                    'num_user_for_reviews', 
                    'budget',
                    'title_year', 
                    'actor_2_facebook_likes', 
                    'imdb_score', 
                    'aspect_ratio',
                    'movie_facebook_likes',
                   ]

for feat in numeric_features:
    print("Az outlierek száma a %s tulajdonságban: %d" % (feat, len(df[ abs(df[feat] - np.mean(df[feat])) >3*np.std(df[feat])])))
    df[feat]= df[feat].apply(lambda x: np.mean(df[feat])+3*np.std(df[feat]) if x > np.mean(df[feat])+3*np.std(df[feat]) else x, 1)
    df[feat]= df[feat].apply(lambda x: np.mean(df[feat])-3*np.std(df[feat]) if x < np.mean(df[feat])-3*np.std(df[feat]) else x, 1)
```
